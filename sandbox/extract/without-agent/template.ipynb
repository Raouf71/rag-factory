## Demo - Extract without agent

```python
# This example shows how to run extraction with configuration from playground without using an agent

from llama_cloud_services import LlamaExtract
from llama_cloud import ExtractConfig

# Initialize client (set LLAMA_CLOUD_API_KEY in your environment)
extractor = LlamaExtract()

# Data Schema and Config
json_data = {
    "dataSchema": {
        "type": "object",
        "properties": {},
        "additionalProperties": False
    },
    "config": {
        "priority": None,
        "extraction_target": "PER_DOC",
        "extraction_mode": "PREMIUM",
        "parse_model": "anthropic-haiku-4.5",
        "extract_model": "openai-gpt-4-1",
        "multimodal_fast_mode": False,
        "system_prompt": None,
        "use_reasoning": False,
        "cite_sources": False,
        "citation_bbox": False,
        "confidence_scores": True,
        "chunk_mode": "PAGE",
        "high_resolution_mode": False,
        "invalidate_cache": False,
        "num_pages_context": None,
        "page_range": None
    }
}

try:
    # Use schema and config from playground
    data_schema = json_data["dataSchema"]
    config = ExtractConfig(**json_data["config"])

    # Extract data directly from document - no agent needed!
    file = "your_document.pdf" # replace with your file name in current directory
    result = extractor.extract(data_schema, config, file)
    print(result.data)
    
except Exception as e:
    print(f"Error: {e}")
```
